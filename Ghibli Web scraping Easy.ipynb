```python
import pandas as pd 
```

```python
url = 'https://ghibliapi.herokuapp.com/films'
```

```python
dataframes = pd.read_json(url)
dataframes.head()
```

```python
i=0
while(i<len(dataframes)):
    print(f"{i+1})",dataframes["title"][i])
    print("(",dataframes["original_title"][i],"):\n")
    print(dataframes["description"][i],"\n")
    i+=1   
```
