```python
import requests #getting the url
from bs4 import BeautifulSoup #extracting data from html 
```

```python
req = requests.get("https://ghibliapi.herokuapp.com/films")
soup = BeautifulSoup(req.text,"html.parser")
text = soup.get_text()
text = text.replace("[","").replace("]","")
text = text.strip()
print(text)
```

```python
#converting string to list
text_list = text.split("},")
print(text_list)
```

```python
#adding '}' to the end of every element of the list
i=0
while (i<len(text_list)-1):
    text_list[i] = text_list[i]+"}"
    i+=1
```

```python
#converting each element into a dictionary
i=0
while(i<len(text_list)):
    if(i!=2 and i!=8):
        Dict = eval(text_list[i].strip()) #eval converts string into dictionary
        print(f"{i+1})",Dict["title"])
        print("(",Dict["original_title"],"):\n")
        print(Dict["description"],"\n")
    i+=1
       
```
